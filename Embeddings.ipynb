```python
import fasttext
```

```python
#sentences with similar words but different contextual meaning..
```

```python
sentences = [
    "I went to the bank to deposit money",
    "We had a picnic by the river bank",
]
```

```python
# Write sentences to a text file
with open('fasttext.txt', 'w') as f:
    for sentence in sentences:
        f.write(sentence + '\n')
```

```python
#for more information you can check out fasttext documentation..
```

```python
# Train the FastText model
model = fasttext.train_unsupervised('fasttext.txt', model='skipgram', dim=50, minCount=1)
```

```python
model
```

```python
# Get all words in the vocabulary
words = model.get_words()
```

```python
words
```

```python
#We can see that the model has only unique words stored in it, and the similar words have same vector..
```

```python
word_embeddings = {}

# Get the embedding for each word
for word in words:
    word_embeddings[word] = model.get_word_vector(word)
```

```python
word_embeddings
```

# CONTEXTUAL EMBEDDING

```python
pip install transformers torch
```

```python
from transformers import BertTokenizer, BertModel
import torch
```

```python
# Load pre-trained BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')
```

```python
# Tokenize the sentences and convert to input tensors
inputs = tokenizer(sentences, return_tensors='pt', padding=True, truncation=True)
```

```python
# Pass the inputs through the BERT model to get contextual embeddings
with torch.no_grad():
    outputs = model(**inputs)
```

```python
# Get the token embeddings for each word in the sentences
token_embeddings = outputs.last_hidden_state
```

```python
token_embeddings
```

```python
# Convert token IDs back to words (tokens)
tokens = [tokenizer.convert_ids_to_tokens(input_id) for input_id in inputs['input_ids']]

# Print the tokens for each sentence
for i, sentence_tokens in enumerate(tokens):
    print(f"\nTokens for sentence {i+1}: {sentence_tokens}")
```

```python
# Find the index of the word "bank" in each sentence
tokens = [tokenizer.convert_ids_to_tokens(input_id) for input_id in inputs['input_ids']]
bank_indices = [[i for i, token in enumerate(sentence_tokens) if token == 'bank'] for sentence_tokens in tokens]
```

```python
# Extract the embeddings for "bank" in both contexts
for i, bank_idx in enumerate(bank_indices):
    print(f"\nEmbedding for 'bank' in sentence {i+1}:")
    for idx in bank_idx:
        print(token_embeddings[i, idx].numpy())  # Convert the tensor to a numpy array for readability
```
